# Bar Chart

```python
months = ['Jan', 'fev', 'Mar', 'Avr']
revenue = [500, 200, 400, 800]
```

```python
plt.figure(figsize=(14, 6))
plt.bar(months, revenue);
plt.title('Monthly Revenue Bar Chart');
plt.xlabel('Months');
plt.ylabel('Revenue');
```

# Subplots and Legends

```python
days = [1, 2, 3, 4, 5]
original_price = [10, 20, 25, 30, 40]
preedicted_price = [8, 11, 21, 25, 28]
```

```python
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.plot(days, preedicted_price, 'r-o', label='Predicted Price')
plt.legend()
plt.title('Plot with Legends')
plt.xlabel('Weekends')
plt.ylabel('Prices')
```

## Subplots

```python
# First subplot - Original Price
plt.subplot(3, 1, 1)
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.title('Original Prices Over Time')
plt.xlabel('Weekends')
plt.ylabel('Prices')
plt.legend()

# Second subplot - Original Price
plt.subplot(3, 1, 2)
plt.plot(days, preedicted_price, 'r-o', label='Predicted Price')
plt.title('Predicted Prices Over Time')
plt.xlabel('Weekends')
plt.ylabel('Prices')
plt.legend()

# Third subplot - Difference between Original & Predicted
price_difference = [o - p for o, p in zip(original_price, preedicted_price)]
plt.subplot(3, 1, 3)
plt.bar(days, preedicted_price, color='purple', label='Price Diference')
plt.title('Difference between Original & Predicted Price')
plt.xlabel('Days')
plt.ylabel('Price Difference')
plt.legend()

plt.tight_layout()
```

## Saving the Figure

```python
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.plot(days, preedicted_price, 'r-o', label='Predicted Price')
plt.legend()
plt.title('Plot with Legends')
plt.xlabel('Weekends')
plt.ylabel('Prices')

plt.savefig('exemple_prediction.png', dpi=300)
```
